### O exercÃ­cio foi retirado da lista de EletrÃ´nica A

Deve-se obter os parÃ¢metros da polarizaÃ§Ã£o DC para um circuito polarizaÃ§Ã£o por divisor de tensÃ£o (PDT) para transistor BJT.




```python
# Pelo circuito temos:
VCC=22
R1=1.8e3
R2=500
RC=600
RE=200
beta=140
VBE=0.7

#Podemos calcular:
VTH = VCC*R2/(R1+R2)
RTH = R1*R2/(R1+R2)
IB = (VTH-VBE)/(RTH+(beta+1)*RE)
IC = beta*IB
IE = (beta+1)*IB

VCE = VCC -IC*RC -IE*RE

#Printando os valores obtidos:
print("TensÃ£o de Thevenin, VTH Ã© =",VTH,"V")
print("Resistor equivalente de Thevenin, RTH Ã© =",RTH,"Î©")
print("Corrente da base, IB Ã© =",IB,"A")
print("Corrente da base, IB Ã© =",(IB*1e6),"ÂµA \n")

print("Corrente do coletor, IC Ã© =",IC)
print("Corrente do coletor, IC Ã© =",(IC*1e3),"mA \n")

print("Corrente do emissor, IE Ã© =",IE)
print("Corrente do emissor, IE Ã© =",(IE*1e3),"mA \n")

print("TensÃ£o entre Emissor e Coletor, VCE Ã© =",VCE,"V")
```

```python
#Calculando outros valores de tensÃ£o
VC = VCC - IC*RC
VE = IE*RE
VB = VTH - IB*RTH

#Printando os valores obtidos:
print("TensÃ£o entre Coletor e Terra, VC Ã© =",VC,"V")
print("TensÃ£o entre Emissor e Terra, VE Ã© =",VE,"V")
print("TensÃ£o entre Base e Terra, VB Ã© =",VB,"V")
```

```python
#Pontos para traÃ§ar a reta de carga,
#ou seja, a resposta do circuito de polarizaÃ§Ã£o
ICsat= VCC/(RC+RE)
VCEcorte = VCC

#Printando os valores obtidos:
print("Corrente de saturaÃ§Ã£o, ou corrente IC mÃ¡xima =",ICsat,"A")
print("Corrente de saturaÃ§Ã£o, ou corrente IC mÃ¡xima =",(ICsat*1e3),"mA \n")

print("TensÃ£o de corte, ou tensÃ£o VCE mÃ¡xima Ã© =",VCEcorte,"V")
```
